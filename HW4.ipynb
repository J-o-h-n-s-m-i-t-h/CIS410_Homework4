```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Read the data from the CSV file into a DataFrame and display the first five rows.
data = pd.read_csv('hw4_exams.csv')
data.head()
```

```python
# 2. Display the basic information for the DataFrame and its columns using the info() method.
data.info()
```

```python
# 3. Display statistical information for the math score, reading score, and writing score columns using the describe() method.
data[['math score', 'reading score', 'writing score']].describe()
```

```python
# 4. Group the data by the race/ethnicity column and display the mean scores.
print(data.groupby('race/ethnicity').mean())
```

```python
# 5. Display a single column as a DataFrame with bracket notation.
data[['math score']]
```

```python
# 6. Display a single column as a Series with bracket notation.
data['math score']
```

```python
# 7. Display a single column as a Series with dot notation.
data.gender
```

```python
# 8. Display only rows for females with a math score greater than or equal to 90.
data[(data.gender == 'female') & (data['math score'] >= 90)]
```

```python
# 9. Use the pandas plot() method to create a box plot.
data.boxplot(column=['math score', 'reading score', 'writing score'])
plt.show()
```

```python
# 10. Group the data by the gender column and calculate the average scores. Then create a bar plot.
gender_avg_scores = data.groupby('gender').mean()
gender_avg_scores.plot(kind='bar')
plt.show()
```

```python
# 11. Use the Seaborn catplot() method to create a count plot.
sns.catplot(x='race/ethnicity', kind='count', data=data)
plt.xticks(rotation=45)  # Rotate x labels for readability
plt.show()
```

```python
# 13. Create a scatter plot comparing writing score with reading score, differentiating by test prep course.
sns.scatterplot(x='reading score', y='writing score', hue='test preparation course', data=data)
plt.show()
```

```python
# 14. Create the same plot with a general method.
plt.scatter(data['reading score'], data['writing score'], c=data['test preparation course'].map({'none': 'red', 'completed': 'blue'}))
plt.show()
```

```python
# 15. Adjust the size of the scatter plot created with the general method.
plt.figure(figsize=(8, 6))  # Adjust size
plt.scatter(data['reading score'], data['writing score'], c=data['test preparation course'].map({'none': 'red', 'completed': 'blue'}))
plt.show()
```
